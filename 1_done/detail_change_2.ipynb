```python
import pandas
import pandas as pd
```

```python
raw_df = pandas.read_excel('01清华大学.xlsx')
```

```python
print(raw_df)
```

```python
edu_list = []
time = []
things = []
for row in raw_df.itertuples():
    temp = str(row[6])
    for line in temp.split('\n'):
        print(line)
        print("#############")
        if len(time) == 0:
            time = line
            # print(time)
            # print(len(time))
        else:
            things = line
            edu_list.append({
                '姓名':row[1],
                '时间':time,
                '教育背景':things
            })
            time = []
            things = []
```

```python
work_list = []
# time = []
# things = []
for row in raw_df.itertuples():
    time = ''
    things = ''
    temp = str(row[7])
    for line in temp.split('\n'):
        # print(line)
        # print("#############")
        if len(time) == 0:
            time = line
            # print(time)
            # print(len(time))
        else:
            things = line
            work_list.append({
                '姓名':row[1],
                '时间':time,
                '工作履历':things
            })
            time = []
            things = []
```

```python
edu_pd = pd.DataFrame(edu_list)
edu_pd.to_excel('edu_pd.xlsx')
work_pd = pd.DataFrame(work_list)
work_pd.to_excel('work_pd.xlsx')
```
